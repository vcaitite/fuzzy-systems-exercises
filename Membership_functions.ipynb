## PART 2 - Creating my own Membership Functions 

```python
def trimf(x,a,b,c):
    y = np.zeros(len(x))
    for i in range(len(x)):
        y[i] = max([min([(x[i]-a)/(b-a), (c-x[i])/(c-b)]), 0])
    return y

def gaussmf(x,c,sigma):
    return e**(-1/2 * ((x-c)/sigma)**2)

def trapmf(x,a,b,c,d):
    y = np.zeros(len(x))
    for i in range(len(x)):
        if x[i] > a and x[i] <= b:
            y[i] = (x[i]-a)/(b-a)
        if x[i] > b and x[i] <= c:
            y[i] = 1
        if x[i] > c and x[i] < d:
            y[i] = (-x[i]+d)/(-c+d)
    return y

def gbellmf(x, a, b, c):
    return 1/(1+abs(((x-c)/a)**(2*b)))

def sigmf(x, c, a):
    return 1/(1 + e**(-a*(x-c)))
    
```

### Triangular Membership Function

```python
x = np.linspace(0, 10, 500)
y1 = trimf(x, 3, 4, 5)
y2 = trimf(x, 2, 4, 7)
y3 = trimf(x, 1, 4, 9)
plt.plot(x, y1, 'b-',x, y2,'r-', x, y3, 'g-')
plt.grid()
legend = plt.legend([r'$y_1$', r'$y_2$', r'$y_3$'])
```

```python
x = np.linspace(0, 10, 500)
y1 = trapmf(x, 2, 3, 7, 9)
y2 = trapmf(x, 3, 4, 6, 8)
y3 = trapmf(x, 4, 5, 5, 7)
plt.plot(x, y1, 'b-',x, y2,'r-', x, y3, 'g-')
plt.grid()
legend = plt.legend([r'$y_1$', r'$y_2$', r'$y_3$'])
```
